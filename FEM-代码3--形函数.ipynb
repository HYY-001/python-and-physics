## 形函数的思想，是让作用域中**对应节点位置的值为1，其他位置的值为0**。

# 一维情况

## 1.1 一维1次形函数

```python
# 一维1次形函数
# f1l: 1 D, left f of 1 order 
def f1l1(x,xl,xr):
    return (x-xr)/(xl-xr)

# f1r: 1 D, righ f of 1 order 
def f1r1(x,xl,xr):
    return (x-xl)/(xr-xl)
```

设定一维1次形函数作用域，并画出形函数

```python
# 这些数字都可以使用变量表示
# 形函数作用域
x,y=np.linspace(0,1),np.linspace(0,1)

# 对于一维，1次性函数
# 左右节点分别为0,1
xl,xr=0,1

# 画出一维一次形函数
plt.plot(x,f1l1(x,xl,xr),label='f1l1')
plt.plot(x,f1r1(x,xl,xr),label='f1r1')
plt.legend()
plt.show()
```

## 1.2 一维2次形函数

```python
# 一维2次形函数
# f1l: 1 D, left f of 2 order
def f1l2(x,xl,xm,xr):
    return (x-xm)*(x-xr)/((xl-xm)*(xl-xr))

# f1m: 1 D, middle f of 2 order
def f1m2(x,xl,xm,xr):
    return (x-xl)*(x-xr)/((xm-xl)*(xm-xr))

# f1r: 1 D, right f of 2 order
def f1r2(x,xl,xm,xr):
    return (x-xl)*(x-xm)/((xr-xl)*(xr-xm))
```

设定一维2次形函数的作用域，并画出该形函数

```python
# 这些数字都可以使用变量表示
# 设定函数作用域
x=np.linspace(0,1)

# 对于一维，2次形函数
# 左右节点分别为0,0.5,1
xl,xm,xr=0,0.5,1

# 画出一维2次形函数
plt.plot(x,f1l2(x,xl,xm,xr),label='f1l2')
plt.plot(x,f1m2(x,xl,xm,xr),label='f1m2')
plt.plot(x,f1r2(x,xl,xm,xr),label='f1r2')
plt.legend()
plt.show()
```

# 2. 二维情况

对于二维1次形函数，可以通过一维1次形函数的组合来构成

## 2.1 二维1次形函数

```python
# y方向用上下，即 up，down表示
# x方向用左右，即left，right表示
# 个人习惯先说上下，再说左右，函数定义根据个人习惯而来

# f2dl1: 2D, down-left f of 1 order
def f2dl1(y,x,yd,yu,xl,xr):
    return f1l1(y,yd,yu)*f1l1(x,xl,xr)

# f2dr1: 2D, down-right f of 1 order
def f2dr1(y,x,yd,yu,xl,xr):
    return f1l1(y,yd,yu)*f1r1(x,xl,xr)

def f2ul1(y,x,yd,yu,xl,xr):
    return f1r1(y,yd,yu)*f1l1(x,xl,xr)

def f2ur1(y,x,yd,yu,xl,xr):
    return f1r1(y,yd,yu)*f1r1(x,xl,xr)
```

定义函数作用域并，画出2D1次形函数

```python
plt.style.use(['default']) # 三维图，还是defaul好看
```

```python
# 这些数字也都可以使用变量表示
x,y=np.linspace(0,1),np.linspace(0,1)
X,Y=np.meshgrid(x,y)

yd,yu=0,1
xl,xr=0,1
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,f2dl1(Y,X,yd,yu,xl,xr),)
ax.plot_surface(X,Y,f2dr1(Y,X,yd,yu,xl,xr),)
ax.plot_surface(X,Y,f2ul1(Y,X,yd,yu,xl,xr),)
ax.plot_surface(X,Y,f2ur1(Y,X,yd,yu,xl,xr),)
ax.set_xlabel('$x$',fontsize=20)
ax.set_ylabel('$y$',fontsize=20)
# plt.colorbar()
plt.show()
```

## 2.2 二维2次形函数

对于二维2次形函数，可以通过一维2次形函数的组合来构成

```python
# y方向用上下，即 up，down表示
# x方向用左右，即left，right表示
# 个人习惯先说上下，再说左右，函数定义根据个人习惯而来

# f2dl2: 2D, down-left f of 1 order
def f2dl2(y,x,yd,ym,yu,xl,xm,xr):
    return f1l2(y,yd,ym,yu)*f1l2(x,xl,xm,xr)

def f2dm2(y,x,yd,ym,yu,xl,xm,xr):
    return f1l2(y,yd,ym,yu)*f1m2(x,xl,xm,xr)

def f2dr2(y,x,yd,ym,yu,xl,xm,xr):
    return f1l2(y,yd,ym,yu)*f1r2(x,xl,xm,xr)

def f2ml2(y,x,yd,ym,yu,xl,xm,xr):
    return f1m2(y,yd,ym,yu)*f1l2(x,xl,xm,xr)

def f2mm2(y,x,yd,ym,yu,xl,xm,xr):
    return f1m2(y,yd,ym,yu)*f1m2(x,xl,xm,xr)

def f2mr2(y,x,yd,ym,yu,xl,xm,xr):
    return f1m2(y,yd,ym,yu)*f1r2(x,xl,xm,xr)

def f2ul2(y,x,yd,ym,yu,xl,xm,xr):
    return f1r2(y,yd,ym,yu)*f1l2(x,xl,xm,xr)

def f2um2(y,x,yd,ym,yu,xl,xm,xr):
    return f1r2(y,yd,ym,yu)*f1m2(x,xl,xm,xr)

def f2ur2(y,x,yd,ym,yu,xl,xm,xr):
    return f1r2(y,yd,ym,yu)*f1r2(x,xl,xm,xr)
```

定义函数作用域并，画出2D2次形函数

```python
# 这些数字也都可以使用变量表示
x,y=np.linspace(0,1),np.linspace(0,1)
X,Y=np.meshgrid(x,y)

yd,ym,yu=0,0.5,1
xl,xm,xr=0,0.5,1
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,f2dl2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2dm2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2dr2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2ml2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2mm2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2mr2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2ul2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2um2(Y,X,yd,ym,yu,xl,xm,xr))
ax.plot_surface(X,Y,f2ur2(Y,X,yd,ym,yu,xl,xm,xr))
# ax.plot_surface(X,Y,Z1)
# ax.plot_surface(X,Y,Z2)
# ax.plot_surface(X,Y,Z3)
plt.show()
```
